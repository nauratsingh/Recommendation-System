```python
import pandas as pd
```

```python
row = pd.read_csv("C:/Users/rawat/Downloads/movie_dataset.txt")
```

```python
row.shape
```

```python
combined_feature = row['genres'] + row['cast'] + row['keywords'] + row['director'] 
```

```python
data = pd.DataFrame(combined_feature)
```

```python
 data.columns = ['combined_feature']
```

```python
data.head()
```

```python
data['combined_feature'].isnull().sum()
```

```python
data.combined_feature.fillna(" ",inplace = True) 
```

### count vectorizing

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit(data['combined_feature'])
```

```python
cv.vocabulary_
```

```python
vector = cv.transform(data['combined_feature'])
```

```python
vector.shape
```

```python
type(vector)
```

```python
vector.toarray() 
```

### cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(vector)
```

```python
cos_sim.shape
```

```python
data_sim =pd.DataFrame(cos_sim)
```

```python
data_sim.head()
```

```python
def to_lower(item) :
    return item.lower()
cols = []
cols = data['title'].apply(to_lower)
data_sim.columns = cols
```

```python
data_sim.head()
```

```python
def predict():
    movie = input("what do you want to see")
    movie= movie.lower()
    ind = data_sim[data_sim[movie] > 0.196][movie].sort_values(ascending = False).index
    print(cols[ind])
```

```python
predict()
```
